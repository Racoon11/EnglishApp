# Описание
--------------------
Web-приложение для изучения английских слов. 

# Требования
-----------------------
## Функциональные требования
### Системные
* Приложение должно сохранять учетные записи
    * логин
    * пароль
    * ник
    * дата регистрации
    * количество выученых слов
    * подневная активность
* Приложение должно сохранять индивидуальный прогресс пользователя<sup>1</sup>

    
    
### Несистемные
* Пользователь должен иметь учетную запись
* Приложение должно требовать пользователя авторизоваться перед началом сессии
* Приложение должно иметь функцию определения уровня владения языка пользователя 
    
    
    
    
    
    
    
<sup>1. сохранять список выученых слов; words that being learnt, количество оставшихся повторение и через какое время их надо повторить; сохранять уровень владения языка; 
    </sup>

NB: Response time
