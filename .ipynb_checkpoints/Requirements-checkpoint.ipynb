# Описание
--------------------
Web-приложение для изучения английских слов. 

# Требования
-----------------------
## Функциональные требования
### Системные

* Приложение должно сохранять учетные записи
    * логин
    * пароль
    * ник
    * дата регистрации
    * количество выученых слов
    * подневная активность
* Приложение должно сохранять индивидуальный прогресс пользователя<sup>1</sup>

    
    
### Несистемные
* Пользователь должен иметь учетную запись
* Приложение должно требовать пользователя авторизоваться перед началом сессии
* Приложение должно иметь функцию определения уровня владения языка пользователя 
* Приложение должно иметь функцию изучения/повторения слов
    * Выбрать правильный перевод (с английского на русский и с русского на английский) из 4 возможных вариантов (возможность кликнуть мышкой на правильный ответ или нажать на цифру на клавиатуре)
    * Восстановление слова по анаграмме
    * Написать слово на английском (дан русский перевод)
* После тренировки приложение должно показывать слова, в которых была совершена хоть одна ошибка
* Слова, в которых была допущена хоть одна ошибка, отправляются на тренировку с сохранение прогресса изучения
* Чтобы изучить слово, нужно повторить их без ошибок четыре раза 
* Приложение должно показывать слова, которые нужно повторить
* Приложение должно показывать список рекомендуемых слов
* Приложение должно иметь функцию добавления новых слов для изучения (вручную)

    
    
    
    
    
    
<sup>1. сохранять список выученых слов; words that being learnt, количество оставшихся повторение и через какое время их надо повторить; сохранять уровень владения языка; 
    </sup>

NB: Response time, add a decent data base with words
