# Описание
--------------------
Web-приложение для изучения английских слов. 

# Требования
-----------------------
## Функциональные требования
### Системные

* Приложение должно сохранять учетные записи
    * логин
    * пароль
    * ник
    * дата регистрации
    * количество выученых слов
    * подневная активность
* Приложение должно сохранять индивидуальный прогресс пользователя<sup>1</sup>

    
    
### Несистемные
* Пользователь должен иметь учетную запись
* Приложение должно требовать пользователя авторизоваться перед началом сессии
* Приложение должно иметь функцию определения уровня владения языка пользователя 
* Приложение должно иметь функцию изучения/повторения слов
    * Выбрать правильный перевод (с английского на русский и с русского на английский) из 4 возможных вариантов (возможность кликнуть мышкой на правильный ответ или нажать на цифру на клавиатуре)
    * Восстановление слова по анаграмме
    * Написать слово на английском (дан русский перевод)
* После тренировки приложение должно показывать слова, в которых была совершена хоть одна ошибка
* Слова, в которых была допущена хоть одна ошибка, отправляются на тренировку с сохранение прогресса изучения
* Чтобы изучить слово, нужно повторить их без ошибок четыре раза 
* Приложение должно показывать слова, которые нужно повторить
* Приложение должно показывать список рекомендуемых слов
* Приложение должно иметь функцию добавления новых слов для изучения (вручную)

    
    
    
    
    
    
<sup>1. сохранять список выученых слов; words that being learnt, количество оставшихся повторение и через какое время их надо повторить; сохранять уровень владения языка; 
    </sup>

NB: Response time, add a decent data base with words

## API contract


```
Request 
  URI: /login
  HTTP Verb: GET
  Body: None
  
Response:
  "login"
 ```

<!-- вот это скорее всего неправильно  -->
```
Request

  HTTP Verb: POST
  Body example:
  {
    "email":"email@smth.em",
    "password":123,
  }
  
Response: 
  HTTP Status:
    200 OK 
    404 NOT FOUND 
```
---


```

  HTTP Verb: GET
  Body: None
  
Response: 
  "registration"

  URI: /registration
  HTTP Verb: POST
  Body example:
  {
    "firstName":"name",
    "lastName" : "last",

  }
  
Response:
```

```
  URI: /registration/email?name={s}
  HTTP Verb: GET
  Body: None
  
Response:

      200 OK "true" если email (name) уже занят
      200 OK "false" иначе
```   

---
```
Request
  URI: /LK
  HTTP Verb: GET
  Body: None
  
Response: 
  HTTP Status:
      200 OK Если user авторизован возвращает его данные
      Body example:
        {
            "id": 1,
            "firstName": "s",
            "lastName": "s",
            "level": "Beginner",
            "date": "Sat Nov 11 14:41:27 GMT+07:00 2023",
            "words": 0,
            "email": "s"
        }
        
Request   
  URI: /LK
  HTTP Verb: PUT
  Body:
  {
      ...
  }
  
Response:
  HTTP Status:
      200 OK Если данные изменены
      Body example:
        {
            "id": 1,
            "firstName": "s",
            "lastName": "s",
            "level": "Beginner",
            "date": "Sat Nov 11 14:41:27 GMT+07:00 2023",
            "words": 0,
            "email": "s"
        }
```
---
```
Request
  URI: /LK/{nick}
  HTTP Verb: GET
  Body: None
  
Response: 
  HTTP Status:
      200 OK Если nick есть в базе
      Body example:
        {
            "id": 2,
            "firstName": "name",
            "lastName": "last",
            "level": "Beginner",
            "date": "Sat Nov 11 16:18:27 GMT+07:00 2023",
            "words": 0,
            "email": "email"
        }
       404 NOT FOUND
```
---
```
Request
  URI: /words
  HTTP Verb: GET
  Body: None
  
Response: 
  HTTP Status:
      200 OK Возвращает список добавленных слов
      Body example:
        [
            {
                "id": 1,
                "wordEng": "word",
                "wordRus": "слово",
                "whenToTrain": 1699694752019,
                "email": "email",
                "county": 0
            },
            {
                "id": 2,
                "wordEng": "word2",
                "wordRus": "слово2",
                "whenToTrain": 1699694949481,
                "email": "email",
                "county": 0
            }
        ]
```
----
```
Request
  URI: /words/{id}
  HTTP Verb: DELETE
  Body: None
  
Response: 
  HTTP Status:
      404 NOT FOUND Слово не в базе
      204 NO CONTENT Слово удалено
```
